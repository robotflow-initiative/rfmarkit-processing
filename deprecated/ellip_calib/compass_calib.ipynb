```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
# os.chdir('/Users/liyutong/projectExchange/stm32-imu-node/python')
df = pd.read_csv('./compass_data_origin.csv', header = None)
df.columns = ['x', 'y', 'z']
```

```python
df
```

```python
mx = np.array(df['x'])
my = np.array(df['y'])
mz = np.array(df['z'])
```

```python
fig = plt.figure(figsize=(16,8))
ax = fig.add_subplot(121, projection='3d')

ax.scatter(mx, my, mz, c=mz)
# 添加坐标轴(顺序是Z, Y, X)
ax.set_zlabel('Z', fontdict={'size': 15, 'color': 'red'})
ax.set_ylabel('Y', fontdict={'size': 15, 'color': 'red'})
ax.set_xlabel('X', fontdict={'size': 15, 'color': 'red'})


ax = fig.add_subplot(122, projection='3d')
ax.scatter(mx, my, mz, c=mz)
# 添加坐标轴(顺序是Z, Y, X)
ax.set_zlabel('Z', fontdict={'size': 15, 'color': 'red'})
ax.set_ylabel('Y', fontdict={'size': 15, 'color': 'red'})
ax.set_xlabel('X', fontdict={'size': 15, 'color': 'red'})
ax.view_init(elev=20,    # 仰角
             azim=45    # 方位角
            )
```

```python
fig = plt.figure(figsize=(24,8))

ax = fig.add_subplot(131)
ax.scatter(mx, my, c=mz)
ax.set_xlabel('X', fontdict={'size': 15, 'color': 'red'})
ax.set_ylabel('Y', fontdict={'size': 15, 'color': 'red'})
ax.set_aspect(1)

ax = fig.add_subplot(132)
ax.scatter(my, mz, c=mz)
ax.set_xlabel('Y', fontdict={'size': 15, 'color': 'red'})
ax.set_ylabel('Z', fontdict={'size': 15, 'color': 'red'})
ax.set_aspect(1)

ax = fig.add_subplot(133)
ax.scatter(mx, mz, c=mz)
ax.set_xlabel('X', fontdict={'size': 15, 'color': 'red'})
ax.set_ylabel('Z', fontdict={'size': 15, 'color': 'red'})
ax.set_aspect(1)
```
